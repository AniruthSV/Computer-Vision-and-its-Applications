```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/drive/MyDrive/CV Images/CV Sample image.jpg')
if image is None:
    print("Error: Image not found.")
    exit()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(10, 4))
plt.title("Histogram of Original Grayscale Image")
plt.hist(gray.ravel(), bins=256, range=[0, 256], color='gray')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
equalized = cv2.equalizeHist(gray)

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_result = clahe.apply(gray)
```

```python
print("Original Grayscale Image:")
cv2_imshow(gray)
print("Histogram Equalized")
cv2_imshow(equalized)
print("CLAHE Result")
cv2_imshow(clahe_result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
