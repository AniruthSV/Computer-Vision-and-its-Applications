```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
image = cv2.imread('/content/drive/MyDrive/CV Images/example.jpg')
a = 1.5
b = 0.9
new_image = cv2.convertScaleAbs(image, alpha=a, beta=b)
print("Original Image:")
cv2_imshow(image)
print("\n")
print("Contrast/Brightness Adjusted image:")
cv2_imshow(new_image)
```
