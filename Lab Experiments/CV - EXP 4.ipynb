```python
import cv2
from google.colab import files
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('      ')
if img is None:
    print("Error: Image not found.")
    exit()
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
avg_filtered=cv2.blur(gray,(5,5))
gaussian_filtered=cv2.GaussianBlur(gray,(5,5),0)
median_filtered=cv2.medianBlur(gray,5)
```

```python
laplacian_filtered=cv2.Laplacian(gray,cv2.CV_64F)
laplacian_filtered=cv2.convertScaleAbs(laplacian_filtered)

sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
sobel_filtered = cv2.convertScaleAbs(cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0))
```

```python
print('\nGray Image\n')
cv2_imshow( gray)
print('\nAverage Filtered Image\n')
cv2_imshow(avg_filtered)
print('\nGaussian Filtered Image\n')
cv2_imshow(gaussian_filtered)
print('\nMedian Filtered Image\n')
cv2_imshow(median_filtered)
print('\nLaplacian Filtered Image\n')
cv2_imshow(laplacian_filtered)
print('\n Sobel Filtered Image\n')
cv2_imshow(sobel_filtered)
```
